# Import the libraries

```python
from pathlib import Path
import tqdm
import torch
import pandas as pd
import numpy as np
from diffusers import StableDiffusionPipeline
from transformers import pipeline, set_seed
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import cv2
```

# Defining Configuration Class

```python
class CFG:
    device = "cuda"
    seed = 42
    generator = torch.Generator(device).manual_seed(seed)
    image_gen_steps = 35
    image_gen_model_id = "stabilityai/stable-diffusion-2"
    image_gen_size = (400,400)
    image_gen_guidance_scale = 9
    prompt_gen_model_id = "gpt2"
    prompt_dataset_size = 6
    prompt_max_length = 12
```

# Image Generation Model

```python
# I couldn't show the token of HuggingFace for the security issues.

"""
we can get the hugging face token from the huggingface: follow this path : click on  "profile" at right top  --> and then on "settings"  --> and
then on "Access Token" and you can create a new token and use that token to run this code.
"""

image_gen_model = StableDiffusionPipeline.from_pretrained(
    CFG.image_gen_model_id, torch_dtype=torch.float16,
    revision="fp16", use_auth_token='Huggingface_authorization_token', guidance_scale=9
)
image_gen_model = image_gen_model.to(CFG.device)
```

# Image Generation Function

```python
def generate_image(prompt, model):
    image = model(
        prompt, num_inference_steps=CFG.image_gen_steps,
        generator=CFG.generator,
        guidance_scale=CFG.image_gen_guidance_scale
    ).images[0]

    image = image.resize(CFG.image_gen_size)
    return image
```

# Model Usage

```python
generate_image("astronaut in space", image_gen_model)
```

```python
generate_image("A man is dancing in the sky with his wife", image_gen_model)
```
